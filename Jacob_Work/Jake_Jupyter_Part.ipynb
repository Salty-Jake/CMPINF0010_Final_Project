```python
import pandas as pd
requests = pd.read_csv("Service Requests.csv")
inspections = pd.read_csv("Inspections.csv")
violations = pd.read_csv("Violations.csv")
requests = requests.sort_values('INSPECT_ID')
inspections = inspections.sort_values('INSPECT_ID')
violations = violations.sort_values('INSPECT_ID')
hood_names = {}
hood_inspections = {}
hood_violations = {}
i = 0
j = 0
k = 0
while i < len(requests)-1:
    if requests.iat[i, 6] not in hood_names:
        hood_names[requests.iat[i, 6]] = 0
    hood_names[requests.iat[i, 6]] += 1
    if requests.iat[i, 2] == inspections.iat[j, 1]:
        if requests.iat[i, 6] not in hood_inspections:
            hood_inspections[requests.iat[i, 6]] = 0
        hood_inspections[requests.iat[i, 6]] += 1
        j+=1
        while requests.iat[i,2] == violations.iat[k, 1]:
            if requests.iat[i, 6] not in hood_violations:
                hood_violations[requests.iat[i, 6]] = 0
            hood_violations[requests.iat[i, 6]] += 1
            k+=1
    i+=1
print(hood_names)
```
