I investigated the safety and quality of housing in Pittsburgh neighborhoods. The metric I used was the number of violations per inspection, with lower counts as positive. I narrowed the data to neighborhoods with at least 100 inspections, because a house being inspected is not necessarily a bad thing. More inspections could mean the housing in the area is staying more up-to-date. It could also mean that the housing market is moving quickly in the area because a lot of the time, the purpose for inspecting a house is to be able to sell it. Neighborhoods with high amounts of inspections, while not having as many violations, would mean that the neighborhood has better housing.


```python
import pandas as pd
requests = pd.read_csv("Service Requests.csv")
inspections = pd.read_csv("Inspections.csv")
violations = pd.read_csv("Violations.csv")
requests = requests.sort_values('INSPECT_ID')
inspections = inspections.sort_values('INSPECT_ID')
violations = violations.sort_values('INSPECT_ID')
hood_names = {}
hood_inspections = {}
hood_violations = {}
i = 0
j = 0
k = 0
while i < len(requests)-1:
    if requests.iat[i, 6] not in hood_names:
        hood_names[requests.iat[i, 6]] = 0
    hood_names[requests.iat[i, 6]] += 1
    if requests.iat[i, 2] == inspections.iat[j, 1]:
        if requests.iat[i, 6] not in hood_inspections:
            hood_inspections[requests.iat[i, 6]] = 0
        hood_inspections[requests.iat[i, 6]] += 1
        j+=1
        while requests.iat[i,2] == violations.iat[k, 1]:
            if requests.iat[i, 6] not in hood_violations:
                hood_violations[requests.iat[i, 6]] = 0
            hood_violations[requests.iat[i, 6]] += 1
            k+=1
    i+=1

print("Number of inspections in each neighborhood: ")
for key, value in hood_inspections.items():
    print(f"{key}: {value}")

print("Number of violations in each neighborhood: ")
for key, value in hood_violations.items():
    print(f"{key}: {value}")
    
violation_ratio = {}
for key, value in hood_inspections.items():
    if key in hood_violations:
        if value > 100:
            violation_ratio[key] = hood_violations.get(key)/value

print("Violation Ratio in Each neighborhood")
for key, value in violation_ratio.items():
    print(f"{key}: {value}")   
```
