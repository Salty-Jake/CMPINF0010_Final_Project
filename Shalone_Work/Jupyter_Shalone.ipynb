```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Transportation 2010 Snap.csv")
df.head()
```

```python
df["PublicTrans"] = df["Commute to Work: Public Transportation (2010)"]
```

```python
top = df.sort_values(by="PublicTrans", ascending=False)
top[["Neighborhood", "PublicTrans"]].head(10)
```

```python
top.head(10).plot(
    x="Neighborhood",
    y="PublicTrans",
    kind="bar",
    title="Top Neighborhoods by Public Transportation Usage (2010)"
)

plt.ylabel("Percentage")
plt.xticks(rotation=45)
plt.show()
```

This analysis looks at how people use public transportation across Pittsburgh neighborhoods. Higher public transportation use may mean fewer people have cars, which can show financial need. This helps compare neighborhoods for our “best neighborhood” project. It also shows that some neighborhoods may have more financial challenges than others.
